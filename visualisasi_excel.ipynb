```python
import plotly.figure_factory as ff
import plotly.graph_objs as go
import numpy as np
def tampilan_vektor_heatmap(Y,colorscale):
  x = Y['koor_x']
  y = Y['koor_y']
  a = Y['Vx']
  b = Y['Vy']
  angle = np.arctan(b / a)
  norm = 5
  u = norm * np.cos(angle)
  v = norm * np.sin(angle)
  f = ff.create_quiver(x, y, u, v, scale = 1, scaleratio = 0.5)

  # misal heatmapnya kurang pas coba geser posisi atau skalanya (persamaan di variabel x, y dalam countour)
  trace1 = f.data[0]
  trace2 = go.Contour(
       x = (x*1.02)-0.5,
       y = (y*1.2)-0.5,
       z = ((a**2)*(b**2))**0.5,
       colorscale = colorscale,
       reversescale = False
   )
  data=[trace1,trace2]
  fig = go.FigureWidget(data)
  fig.show()
```

```python
tampilan_vektor_heatmap(df,'plasma')
```
