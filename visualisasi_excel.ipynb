```python
import plotly.graph_objects as go
from plotly.figure_factory import create_quiver

# koordinat x = x
# koor y = y
# magnitude sumbu x = Vx
# magnitude sumbu y = Vy
# tipe data masing2 array, bisa dubah di excelnya
# a = [ 1, 2, 3]
# 3 parameter, file yg mau divisualisasi, skala panah, skala kepala panah

def tampilan_vektor(Y,scale,arrow):
  x = - Y['koor_x']
  y = Y['koor_y']
  u =  - Y['Vx']
  v = Y['Vy']

  fig = create_quiver(x,y,u,v, scale=scale, arrow_scale=arrow)
  
  fig.show()
```

```python
# masukkan parameter di dalam kurung fungsi sesuai urutan

tampilan_vektor(df,50,0.5)
```
